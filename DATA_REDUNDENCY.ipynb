```python
# ==========================================
# CodeAlpha Task 1
# Data Redundancy Removal System with Graphs
# Dataset: marketing_campaign.csv
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('/content/marketing_campaign.csv')

# Show first 5 rows
print("Original Dataset:\n")
print(df.head())

# Total records before cleaning
before = len(df)

# Find duplicate rows
duplicates = df[df.duplicated()]

print("\nDuplicate Records Found:\n")
print(duplicates)

# Remove duplicates
cleaned_df = df.drop_duplicates()

# Total records after cleaning
after = len(cleaned_df)

# Save cleaned dataset
cleaned_df.to_csv('/content/cleaned_marketing_campaign.csv', index=False)

print("\nTotal Records Before Cleaning:", before)
print("Total Records After Cleaning:", after)
print("Duplicates Removed Successfully!")

# ==========================================
# GRAPH 1 - Before vs After Cleaning
# ==========================================

plt.figure(figsize=(6,5))

labels = ['Before Cleaning', 'After Cleaning']
values = [before, after]

sns.barplot(x=labels, y=values, palette='Set2')

plt.title('Data Redundancy Removal')
plt.ylabel('Number of Records')

for i, v in enumerate(values):
    plt.text(i, v + 5, str(v), ha='center', fontsize=12)

plt.show()

# ==========================================
# GRAPH 2 - Duplicate vs Unique Records
# ==========================================

duplicate_count = before - after
unique_count = after

sizes = [duplicate_count, unique_count]
labels = ['Duplicate Records', 'Unique Records']
colors = ['red', 'green']

plt.figure(figsize=(6,6))

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Duplicate vs Unique Records')

plt.show()

# ==========================================
# Download cleaned dataset
# ==========================================

from google.colab import files
files.download('/content/cleaned_marketing_campaign.csv')
```
